##BILLBOARD WEEKLY MUSIC RANK DATASET (QUESTION 2)

```python
import pandas as pd
weekly_music_data = pd.read_csv('billboard.csv')
weekly_music_data
```

```python
weekly_music_data.fillna(0,inplace = True)
weekly_music_data
```

```python
weekly_music_data.columns
```

```python
melted= pd.melt(weekly_music_data,id_vars=['artiste', 'track', 'genre', 'date.entered', 'date.peaked'],value_vars=['x1st.week',
       'x2nd.week', 'x3rd.week', 'x4th.week', 'x5th.week', 'x6th.week','x7th.week', 'x8th.week', 'x9th.week', 'x10th.week', 'x11th.week',
       'x12th.week', 'x13th.week', 'x14th.week', 'x15th.week', 'x16th.week','x17th.week', 'x18th.week', 'x19th.week', 'x20th.week', 'x21st.week',
       'x22nd.week', 'x23rd.week', 'x24th.week', 'x25th.week', 'x26th.week','x27th.week', 'x28th.week', 'x29th.week', 'x30th.week', 'x31st.week',
       'x32nd.week', 'x33rd.week', 'x34th.week', 'x35th.week', 'x36th.week','x37th.week', 'x38th.week', 'x39th.week', 'x40th.week', 'x41st.week',
       'x42nd.week', 'x43rd.week', 'x44th.week', 'x45th.week', 'x46th.week','x47th.week', 'x48th.week', 'x49th.week', 'x50th.week', 'x51st.week',
       'x52nd.week', 'x53rd.week', 'x54th.week', 'x55th.week', 'x56th.week','x57th.week', 'x58th.week', 'x59th.week', 'x60th.week', 'x61st.week',
       'x62nd.week', 'x63rd.week', 'x64th.week', 'x65th.week', 'x66th.week','x67th.week', 'x68th.week', 'x69th.week', 'x70th.week', 'x71st.week',
       'x72nd.week', 'x73rd.week', 'x74th.week', 'x75th.week', 'x76th.week'],var_name='week',value_name='rank')
melted
```

```python
melted.info()
```

```python
melted['week'] = melted['week'].str.extract('(\d+)').astype(int)
```

```python
melted
```

```python
melted['date.entered']=pd.to_datetime(melted['date.entered'],format='%Y-%m-%d')
melted['date.peaked']=pd.to_datetime(melted['date.peaked'],format='%Y-%m-%d')
melted.info()
```

```python
date_sub=melted['date.peaked']-melted['date.entered']
date_sub
```

```python
peak=pd.DataFrame(date_sub,columns=['entered_to_peak'])
peak
```

```python
peak.info()
```

```python
peak=peak['entered_to_peak'].dt.days
```

```python
peak
```

```python
peak=pd.DataFrame(peak,columns=['entered_to_peak'])
peak
```

```python
final_weekly_music_data = pd.concat([melted,peak],axis=1)
final_weekly_music_data
```

```python
f1=final_weekly_music_data[final_weekly_music_data.genre=='Rock' ]
f1
```

```python
f2=f1[f1.entered_to_peak== 35]
f2
```
